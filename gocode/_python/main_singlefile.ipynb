```python
LOCAL_IPERF_TESTS = 100
local_iperf_oauth_tests = []
local_iperf_no_oauth_tests = []
local_iperf_improved_oauth_tests = []

class iperf:
	def __init__(self):
		self.bits_per_second = []
		self.total_bytes = []
		self.testNumber = 0

	def addTest(self, bits_ps, bytes):
		self.bits_per_second.append(bits_ps)
		self.total_bytes[self.testNumber] += bytes

		if self.totalTests() % 10 == 0:
			self.testNumber += 1


	def get_bits_per_second(self):
		return self.bits_per_second

	def getTotalBytes(self):
		return self.total_bytes

	def totalTests(self):
		return len(self.bits_per_second)



local_iperf_no_oauth_results_bits_ps = []
local_iperf_oauth_results_bits_ps = []
local_iperf_no_oauth_results_bytes = []
local_iperf_oauth_results_bytes = []
local_iperf_improved_oauth_results_bytes = []
local_iperf_improved_oauth_results_bits_ps = []
oauth = iperf()
no_oauth = iperf()
improved_oauth = iperf()


with open('./performance/iperf_local/ue_iperf_oauth_local0.json', 'r') as f:
	json_parse = json.load(f)
	for item in json_parse["intervals"]:
		for stream in item["streams"]:
			oauth.addTest(stream["bytes"], stream["bits_per_second"])
	#print(local_iperf_oauth_tests[i]["end"]["sum_received"]["bits_per_second"])

with open('./performance/iperf_local/ue_iperf_no_oauth_local0.json', 'r') as f:
	json_parse = json.load(f)
	for item in json_parse["intervals"]:
		for stream in item["streams"]:
			no_oauth.addTest(stream["bytes"], stream["bits_per_second"])
	#print(local_iperf_no_oauth_tests[i]["end"]["sum_received"]["bits_per_second"])

with open('./performance/iperf_local/ue_iperf_improved_oauth_local0.json', 'r') as f:
	json_parse = json.load(f)
	for item in json_parse["intervals"]:
		for stream in item["streams"]:
			improved_oauth.addTest(stream["bytes"], stream["bits_per_second"])
	#print(local_iperf_no_oauth_tests[i]["end"]["sum_received"]["bits_per_second"])

# for i in range(LOCAL_IPERF_TESTS):
# 	for item in local_iperf_oauth_tests[i]["intervals"]:
# 		for stream in item["streams"]:
# 			print(type(stream))
# 			print(stream["bytes"])
# 			oauth.addTest(stream["bytes"], stream["bits_per_second"])
# 		oauth = iperf(local_iperf_oauth_tests[i]["end"]["sum_received"]["bits_per_second"], local_iperf_oauth_tests[i]["end"]["sum_received"]["bytes"])

# 		no_oauth = iperf(local_iperf_no_oauth_tests[i]["end"]["sum_received"]["bits_per_second"],
# 					local_iperf_no_oauth_tests[i]["end"]["sum_received"]["bytes"])

# 		improved_oauth = iperf(local_iperf_improved_oauth_tests[i]["end"]["sum_received"]["bits_per_second"],
# 							local_iperf_improved_oauth_tests[i]["end"]["sum_received"]["bytes"])

# 	local_iperf_no_oauth_results_bits_ps.append(no_oauth.bits_per_second)
# 	local_iperf_no_oauth_results_bytes.append(no_oauth.total_bytes)

# 	local_iperf_oauth_results_bits_ps.append(oauth.bits_per_second)
# 	local_iperf_oauth_results_bytes.append(oauth.total_bytes)

# 	local_iperf_improved_oauth_results_bits_ps.append(improved_oauth.bits_per_second)
# 	local_iperf_improved_oauth_results_bytes.append(improved_oauth.total_bytes)


# print(local_iperf_oauth_results_bytes)
# print(local_iperf_oauth_results_bits_ps)
# print(local_iperf_no_oauth_results_bytes)
# print(local_iperf_no_oauth_results_bits_ps)
```

# ------------------- UE CONNECTION STRESS TEST GRAPHS ------------------------

```python


def calculate_time_between(date1, date2):
	date1 = datetime.strptime(date1, "%Y-%m-%d %H:%M:%S.%f")
	date2 = datetime.strptime(date2, "%Y-%m-%d %H:%M:%S.%f")
	if date1 >= date2:
		return (date1 - date2).total_seconds()
	else:
		return (date2 - date1).total_seconds()


oauth_data = []
no_oauth_data = []

with open('./performance/ue_connection_timings/oauth.csv', 'r') as f:
	oauth = csv.DictReader(f)
	for row in oauth:
		row['diff'] = calculate_time_between(row['Start Time'], row['End Time'])
		oauth_data.append(row)


with open('./performance/ue_connection_timings/no_oauth.csv', 'r') as f:
	no_oauth = csv.DictReader(f)
	for row in no_oauth:
		row['diff'] = calculate_time_between(row['Start Time'], row['End Time'])
		no_oauth_data.append(row)

oauth_diff = []
no_oauth_diff = []
for row in oauth_data:
	oauth_diff.append(row['diff'])

for row in no_oauth_data:
	no_oauth_diff.append(row['diff'])

ue_init_ci_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(oauth_diff)-1,
                                 loc=np.mean(oauth_diff), scale=st.sem(oauth_diff))

ue_init_ci_no_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(no_oauth_diff)-1,
                                                loc=np.mean(no_oauth_diff), scale=st.sem(no_oauth_diff))
```

```python
fig, ax = plt.subplots()

labels = ["free5gc", "free5gc w/OAuth"]
x = np.arange(len(labels))  # the label locations
#width = 0.35  # the width of the bars

yerr_oauth = [ue_init_ci_oauth[1] - ue_init_ci_oauth[0]]
yerr_no_oauth = [ue_init_ci_no_oauth[1] - ue_init_ci_no_oauth[0]]

print(yerr_no_oauth)
print(yerr_oauth)

print([avg(oauth_diff), avg(no_oauth_diff)])

ax.bar(0, [avg(no_oauth_diff)], capsize=10, yerr=yerr_no_oauth)
ax.bar(1, [avg(oauth_diff)], capsize=10, yerr=yerr_oauth)

ax.set_title("UE Setup Time")
ax.set_ylabel("Average Time in Seconds")
#ax.set_xlabel("Test Number")
ax.set_xticks([0, 1])
ax.set_xticklabels(labels)
fig.set_figwidth(3)
fig.set_figheight(3)

plt.savefig("ue_setup_time.pdf")
```

```python
oauth_data = []
no_oauth_data = []
improved_oauth_data = []

with open('./performance/ue_connection_timings/singular/oauth.csv', 'r') as f:
	oauth = csv.DictReader(f)
	for row in oauth:
		row['diff'] = calculate_time_between(row['Start Time'], row['End Time'])
		oauth_data.append(row)


with open('./performance/ue_connection_timings/singular/no_oauth.csv', 'r') as f:
	no_oauth = csv.DictReader(f)
	for row in no_oauth:
		row['diff'] = calculate_time_between(row['Start Time'], row['End Time'])
		no_oauth_data.append(row)

with open('./performance/ue_connection_timings/singular/improved_oauth.csv', 'r') as f:
	improved_oauth = csv.DictReader(f)
	for row in improved_oauth:
		row['diff'] = calculate_time_between(row['Start Time'], row['End Time'])
		improved_oauth_data.append(row)

oauth_diff = []
no_oauth_diff = []
improved_oauth_diff = []

for row in oauth_data:
	oauth_diff.append(row['diff'])

for row in no_oauth_data:
	no_oauth_diff.append(row['diff'])

for row in improved_oauth_data:
	improved_oauth_diff.append(row['diff'])

ue_init_ci_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(oauth_diff)-1,
                                 loc=np.mean(oauth_diff), scale=st.sem(oauth_diff))

ue_init_ci_no_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(no_oauth_diff)-1,
                                    loc=np.mean(no_oauth_diff), scale=st.sem(no_oauth_diff))

ue_init_ci_improved_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(improved_oauth_diff)-1,
                                          loc=np.mean(improved_oauth_diff), scale=st.sem(improved_oauth_diff))
```

```python
fig, ax = plt.subplots()

labels = ["free5gc", "free5gc\nw/OAuth", "free5gc\nw/ improved OAuth"]
x = np.arange(len(labels))  # the label locations
#width = 0.35  # the width of the bars

yerr_oauth = [ue_init_ci_oauth[1] - ue_init_ci_oauth[0]]
yerr_no_oauth = [ue_init_ci_no_oauth[1] - ue_init_ci_no_oauth[0]]
yerr_improved_oauth = [
    ue_init_ci_improved_oauth[1] - ue_init_ci_improved_oauth[0]]

print(yerr_no_oauth)
print(yerr_oauth)
print(yerr_improved_oauth)

print("Average time to register UE no OAuth: " + str(avg(no_oauth_diff)))
print("Average time to register UE w/ OAuth: " + str(avg(oauth_diff)))
print("Average time to register UE w/ improved OAuth: " +
      str(avg(improved_oauth_diff)))

def percent_diff(num1, num2):
    if num1 == num2:
        return 100.0
    try:
        return (abs(num1 - num2) / num2) * 100.0
    except ZeroDivisionError:
        return 0


print("Adding OAuth causes a " + str(avg(oauth_diff) - avg(no_oauth_diff)
                                     ) + " performance difference which is an increase of " + str(percent_diff(avg(oauth_diff), avg(no_oauth_diff))) + "%")
print("Improving OAuth causes a " + str(avg(improved_oauth_diff) - avg(oauth_diff)
                                        ) + " performance difference which is an increase of " + str(percent_diff(avg(improved_oauth_diff), avg(oauth_diff))) + "%")

print([avg(oauth_diff), avg(no_oauth_diff), avg(improved_oauth_diff)])

ax.bar(0, [avg(no_oauth_diff)], capsize=10, yerr=yerr_no_oauth)
ax.bar(1, [avg(oauth_diff)], capsize=10, yerr=yerr_oauth)
ax.bar(2, [avg(improved_oauth_diff)], capsize=10, yerr=yerr_improved_oauth)

ax.set_title("UE Setup Time")
ax.set_ylabel("Average Time in Seconds")
#ax.set_xlabel("Test Number")
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(labels)
#fig.set_figwidth(3)
#fig.set_figheight(3)

plt.savefig("ue_setup_time_single_use.pdf")
```
