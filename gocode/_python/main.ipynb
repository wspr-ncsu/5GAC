```python
import matplotlib.pyplot as plt
import json
import csv
import csv
from datetime import datetime
import bitmath as b
```

```python
LOCAL_IPERF_TESTS = 10
local_iperf_oauth_tests = []
local_iperf_no_oauth_tests = []
local_iperf_improved_oauth_tests = []

class iperf:
	def __init__(self, bits_ps, bytes):
		self.bits_per_second = bits_ps
		self.total_bytes = bytes


local_iperf_no_oauth_results_bits_ps = []
local_iperf_oauth_results_bits_ps = []
local_iperf_no_oauth_results_bytes = []
local_iperf_oauth_results_bytes = []
local_iperf_improved_oauth_results_bytes = []
local_iperf_improved_oauth_results_bits_ps = []

for i in range(LOCAL_IPERF_TESTS):
	with open('./performance/iperf_local/ue_iperf_oauth_local' + str(i) + '.json', 'r') as f:
		local_iperf_oauth_tests.append(json.load(f))
		#print(local_iperf_oauth_tests[i]["end"]["sum_received"]["bits_per_second"])

	with open('./performance/iperf_local/ue_iperf_no_oauth_local' + str(i) + '.json', 'r') as f:
		local_iperf_no_oauth_tests.append(json.load(f))
		#print(local_iperf_no_oauth_tests[i]["end"]["sum_received"]["bits_per_second"])

	with open('./performance/iperf_local/ue_iperf_improved_oauth_local' + str(i) + '.json', 'r') as f:
		local_iperf_improved_oauth_tests.append(json.load(f))
		#print(local_iperf_no_oauth_tests[i]["end"]["sum_received"]["bits_per_second"])

	oauth = iperf(local_iperf_oauth_tests[i]["end"]["sum_received"]["bits_per_second"], local_iperf_oauth_tests[i]["end"]["sum_received"]["bytes"])

	no_oauth = iperf(local_iperf_no_oauth_tests[i]["end"]["sum_received"]["bits_per_second"],
	              local_iperf_no_oauth_tests[i]["end"]["sum_received"]["bytes"])

	improved_oauth = iperf(local_iperf_improved_oauth_tests[i]["end"]["sum_received"]["bits_per_second"],
                        local_iperf_improved_oauth_tests[i]["end"]["sum_received"]["bytes"])

	local_iperf_no_oauth_results_bits_ps.append(no_oauth.bits_per_second)
	local_iperf_no_oauth_results_bytes.append(no_oauth.total_bytes)

	local_iperf_oauth_results_bits_ps.append(oauth.bits_per_second)
	local_iperf_oauth_results_bytes.append(oauth.total_bytes)

	local_iperf_improved_oauth_results_bits_ps.append(improved_oauth.bits_per_second)
	local_iperf_improved_oauth_results_bytes.append(improved_oauth.total_bytes)


print(local_iperf_oauth_results_bytes)
print(local_iperf_oauth_results_bits_ps)
print(local_iperf_no_oauth_results_bytes)
print(local_iperf_no_oauth_results_bits_ps)
```

```python


def calculate_time_between(date1, date2):
	offset1 = -5
	offset2 = -5
	if len(date1) == 24: # It's the time from UE RAN SIM
		offset1 = len(date1) - 1
	if len(date2) == 24: # It's the time from UE RAN SIM
		offset2 = len(date2) - 1

	print("len: " + str(len(date1)) + " " + str(len(date2)) + " offset: " + str(offset1) + " " + str(offset2))
	print(date1[:offset1])

	date1 = datetime.strptime(date1[:offset1], "%Y-%m-%dT%H:%M:%S.%f")
	date2 = datetime.strptime(date2[:offset2], "%Y-%m-%dT%H:%M:%S.%f")
	if date1 >= date2:
		return (date1 - date2).total_seconds()
	else:
		return (date2 - date1).total_seconds()


oauth_data = []
no_oauth_data = []

with open('./performance/ue_connection_timings/singular/oauth.csv', 'r') as f:
	oauth = csv.DictReader(f)
	for row in oauth:
		row['diff'] = calculate_time_between(
			row['AKA Timing Start'], row['AKA Timing end'])
		oauth_data.append(row)


with open('./performance/ue_connection_timings/singular/no_oauth.csv', 'r') as f:
	no_oauth = csv.DictReader(f)
	for row in no_oauth:
		row['diff'] = calculate_time_between(
			row['AKA Timing Start'], row['AKA Timing end'])
		no_oauth_data.append(row)

oauth_diff = []
no_oauth_diff = []
for row in oauth_data:
	oauth_diff.append(row['diff'])

for row in no_oauth_data:
	no_oauth_diff.append(row['diff'])

ue_init_ci_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(oauth_diff)-1,
                                 loc=np.mean(oauth_diff), scale=st.sem(oauth_diff))

ue_init_ci_no_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(no_oauth_diff)-1,
                                                loc=np.mean(no_oauth_diff), scale=st.sem(no_oauth_diff))
```

```python
oauth_data = []
no_oauth_data = []
improved_oauth_data = []

oauth_data_pdu_session_setup = []
no_oauth_data_pdu_session_setup = []
improved_oauth_data_pdu_session_setup = []

oauth_data_pdu_session_release = []
no_oauth_data_pdu_session_release = []
improved_oauth_data_pdu_session_release = []

oauth_data_deregistration = []
no_oauth_data_deregistration = []
improved_oauth_data_deregistration = []

with open('./performance/ue_connection_timings/singular/oauth.csv', 'r') as f:
	oauth = csv.DictReader(f)
	for row in oauth:
		row['diff_aka'] = calculate_time_between(
			row['AKA Timing Start'], row['AKA Timing end'])
		oauth_data.append(row)

		row['diff_pdu_start'] = calculate_time_between(
			row['PDU Session Establishment Start'], row['PDU Session Establishment End'])
		oauth_data_pdu_session_setup.append(row)

		row['diff_pdu_release'] = calculate_time_between(
			row['PDU Session Release Start'], row['PDU Session Release End'])
		oauth_data_pdu_session_release.append(row)

		row['diff_deregistration'] = calculate_time_between(
			row['Deregistration Start'], row['Deregistration End'])
		oauth_data_deregistration.append(row)


with open('./performance/ue_connection_timings/singular/no_oauth.csv', 'r') as f:
	no_oauth = csv.DictReader(f)
	for row in no_oauth:
		row['diff_aka'] = calculate_time_between(
			row['AKA Timing Start'], row['AKA Timing end'])
		no_oauth_data.append(row)

		row['diff_pdu_start'] = calculate_time_between(
			row['PDU Session Establishment Start'], row['PDU Session Establishment End'])
		no_oauth_data_pdu_session_setup.append(row)

		row['diff_pdu_release'] = calculate_time_between(
			row['PDU Session Release Start'], row['PDU Session Release End'])
		no_oauth_data_pdu_session_release.append(row)

		row['diff_deregistration'] = calculate_time_between(
			row['Deregistration Start'], row['Deregistration End'])
		no_oauth_data_deregistration.append(row)

with open('./performance/ue_connection_timings/singular/improved_oauth.csv', 'r') as f:
	improved_oauth = csv.DictReader(f)
	for row in improved_oauth:
		row['diff_aka'] = calculate_time_between(
			row['AKA Timing Start'], row['AKA Timing end'])
		improved_oauth_data.append(row)

		row['diff_pdu_start'] = calculate_time_between(
			row['PDU Session Establishment Start'], row['PDU Session Establishment End'])
		improved_oauth_data_pdu_session_setup.append(row)

		row['diff_pdu_release'] = calculate_time_between(
			row['PDU Session Release Start'], row['PDU Session Release End'])
		improved_oauth_data_pdu_session_release.append(row)

		row['diff_deregistration'] = calculate_time_between(
			row['Deregistration Start'], row['Deregistration End'])
		improved_oauth_data_deregistration.append(row)




oauth_diff = {}
oauth_diff["aka"] = []
oauth_diff["pdu_start"] = []
oauth_diff["pdu_release"] = []
oauth_diff["deregistration"] = []

no_oauth_diff = {}
no_oauth_diff["aka"] = []
no_oauth_diff["pdu_start"] = []
no_oauth_diff["pdu_release"] = []
no_oauth_diff["deregistration"] = []

improved_oauth_diff = {}
improved_oauth_diff["aka"] = []
improved_oauth_diff["pdu_start"] = []
improved_oauth_diff["pdu_release"] = []
improved_oauth_diff["deregistration"] = []

for row in oauth_data:
	oauth_diff["aka"].append(row['diff_aka'])
	oauth_diff["pdu_start"].append(row['diff_pdu_start'])
	oauth_diff["pdu_release"].append(row['diff_pdu_release'])
	oauth_diff["deregistration"].append(row['diff_deregistration'])

for row in no_oauth_data:
	no_oauth_diff["aka"].append(row['diff_aka'])
	no_oauth_diff["pdu_start"].append(row['diff_pdu_start'])
	no_oauth_diff["pdu_release"].append(row['diff_pdu_release'])
	no_oauth_diff["deregistration"].append(row['diff_deregistration'])

for row in improved_oauth_data:
	improved_oauth_diff["aka"].append(row['diff_aka'])
	improved_oauth_diff["pdu_start"].append(row['diff_pdu_start'])
	improved_oauth_diff["pdu_release"].append(row['diff_pdu_release'])
	improved_oauth_diff["deregistration"].append(row['diff_deregistration'])

def get_ci_intervals(diffs):
	result = {}

	result["aka"] = st.t.interval(alpha=CONF_INTERVAL, df=len(diffs["aka"])-1,
                               loc=np.mean(diffs["aka"]), scale=st.sem(diffs["aka"]))
	result["pdu_start"] = st.t.interval(alpha=CONF_INTERVAL, df=len(diffs["pdu_start"])-1,
                                             loc=np.mean(diffs["pdu_start"]), scale=st.sem(diffs["pdu_start"]))
	result["pdu_release"] = st.t.interval(alpha=CONF_INTERVAL, df=len(diffs["pdu_release"])-1,
                                               loc=np.mean(diffs["pdu_release"]), scale=st.sem(diffs["pdu_release"]))
	result["deregistration"] = st.t.interval(alpha=CONF_INTERVAL, df=len(diffs["deregistration"])-1,
                                          loc=np.mean(diffs["deregistration"]), scale=st.sem(diffs["deregistration"]))

	return result


ue_ci_oauth = get_ci_intervals(oauth_diff)
ue_ci_no_oauth = get_ci_intervals(no_oauth_diff)
ue_ci_improved_oauth = get_ci_intervals(improved_oauth_diff)

ue_init_ci_no_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(no_oauth_diff["aka"])-1,
                                    loc=np.mean(no_oauth_diff["aka"]), scale=st.sem(no_oauth_diff["aka"]))

ue_init_ci_improved_oauth = st.t.interval(alpha=CONF_INTERVAL, df=len(improved_oauth_diff["aka"])-1,
                                          loc=np.mean(improved_oauth_diff["aka"]), scale=st.sem(improved_oauth_diff["aka"]))
```
